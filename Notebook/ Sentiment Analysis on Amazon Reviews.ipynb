## Problem Statement: Amazon Alexa Review Sentiment Analysis

- The rapid growth of voice-based virtual assistants like Amazon Alexa has led to an enormous amount of user feedback in the form of reviews. These reviews provide valuable insights into user satisfaction, product usability, and overall performance. However, manually analyzing thousands of such reviews is time-consuming and inefficient.

- The goal of this project is to develop a Natural Language Processing (NLP) model that can automatically classify Amazon Alexa customer reviews as positive or negative based on their text content.
This sentiment analysis system will help Amazon and other stakeholders understand customer opinions, identify areas of improvement, and enhance user experience through data-driven decisions.

## Objectives

- Collect and preprocess Amazon Alexa review data (cleaning, tokenization, stop-word removal, etc.).

- Convert text into numerical features using methods like TF-IDF, n-grams, or word embeddings.

- Train and evaluate machine learning models such as Logistic Regression, Na√Øve Bayes, or SVM for sentiment classification.

- Assess model performance using metrics like accuracy, precision, recall, and F1-score.

- Visualize results and interpret the most influential words affecting sentiment

## Preprocessing and Modeling
- To build the corpus from the 'verified_reviews' we perform the following -

- Replace any non-alphabet characters with a space
- Covert to lower case and split into words
- Iterate over the individual words, and if it is not a stopword, then add the stemmed form of the word to the corpus

```python
corpus = []
stemmer = PorterStemmer()
for i in range(0, data.shape[0]):
  review = re.sub('[^a-zA-Z]', ' ', data.iloc[i]['verified_reviews'])
  review = review.lower().split()
  review = [stemmer.stem(word) for word in review if not word in STOPWORDS]
  review = ' '.join(review)
  corpus.append(review)
```

Using Count Vectorizer to create bag of words

```python
cv = CountVectorizer(max_features = 2500)

#Storing independent and dependent variables in X and y
X = cv.fit_transform(corpus).toarray()
y = data['feedback'].values
```

```python
#Saving the Count Vectorizer
import os
import pickle

# Define target path
filename = 'Models/countVectorizer.pkl'

# Ensure directory 'Models' exists
os.makedirs(os.path.dirname(filename), exist_ok=True)

# Save your CountVectorizer
pickle.dump(cv, open(filename, 'wb'))
```

- Checking the shape of X and y

```python
print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
```

- Splitting data into a train and test set with 30% data for testing.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 15)

print(f"X train: {X_train.shape}")
print(f"y train: {y_train.shape}")
print(f"X test: {X_test.shape}")
print(f"y test: {y_test.shape}")
```

```python
print(f"X train max value: {X_train.max()}")
print(f"X test max value: {X_test.max()}")
```

- We'll scale X_train and X_test so that all values are between 0 and 1.

```python
scaler = MinMaxScaler()

X_train_scl = scaler.fit_transform(X_train)
X_test_scl = scaler.transform(X_test)
```

```python
#Saving the scaler model
pickle.dump(scaler, open('Models/scaler.pkl', 'wb'))
```

### Random Forest

```python
#Fitting scaled X_train and y_train on Random Forest Classifier
model_rf = RandomForestClassifier()
model_rf.fit(X_train_scl, y_train)
```

```python
#Accuracy of the model on training and testing data
 
print("Training Accuracy :", model_rf.score(X_train_scl, y_train))
print("Testing Accuracy :", model_rf.score(X_test_scl, y_test))
```

```python
#Predicting on the test set
y_preds = model_rf.predict(X_test_scl)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Assuming you already have y_test (true labels) and y_pred (model predictions)
y_pred = model_rf.predict(X_test)

# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Create and plot the confusion matrix display
cm_display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model_rf.classes_)
cm_display.plot(cmap=plt.cm.Blues)
plt.show()
```

### K-fold cross-validation

```python
accuracies = cross_val_score(estimator = model_rf, X = X_train_scl, y = y_train, cv = 10)

print("Accuracy :", accuracies.mean())
print("Standard Variance :", accuracies.std())
```

- Applying grid search to get the optimal parameters on random forest

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier

# Example model
model_rf = RandomForestClassifier(random_state=42)

# Define parameter grid
params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Define cross-validation strategy
cv_object = StratifiedKFold(n_splits=2)

# Set up GridSearchCV
grid_search = GridSearchCV(
    estimator=model_rf,
    param_grid=params,
    cv=cv_object,
    verbose=0,
    return_train_score=True,
    n_jobs=-1
)

# Fit the model
grid_search.fit(X_train_scl, y_train.ravel())

# Best hyperparameters and best model
print("Best Parameters:", grid_search.best_params_)
best_rf = grid_search.best_estimator_
```

```python
#Getting the best parameters from the grid search


print("Best Parameter Combination : {}".format(grid_search.best_params_))
```

```python
print("Cross validation mean accuracy on train set : {}".format(grid_search.cv_results_['mean_train_score'].mean()*100))
print("Cross validation mean accuracy on test set : {}".format(grid_search.cv_results_['mean_test_score'].mean()*100))
print("Accuracy score for test set :", accuracy_score(y_test, y_preds))
```
